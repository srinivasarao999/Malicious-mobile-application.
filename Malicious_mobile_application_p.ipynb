```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import confusion_matrix
# Function to plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Goodware', 'Malware'], yticklabels=['Goodware', 'Malware'])
    plt.title(f'Confusion Matrix for {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()
```

```python
# 3. Support Vector Machine (SVM)
svm = SVC(probability=True)
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)
print("Support Vector Machine (SVM)")
print("Accuracy:", accuracy_score(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))
```

```python
plot_confusion_matrix(y_test, y_pred_svm, 'SVM')
```

```python
# 5. Random Forest
random_forest = RandomForestClassifier()
random_forest.fit(X_train, y_train)
y_pred_random_forest = random_forest.predict(X_test)
print("Random Forest")
print("Accuracy:", accuracy_score(y_test, y_pred_random_forest))
print(classification_report(y_test, y_pred_random_forest))
```

```python
plot_confusion_matrix(y_test, y_pred_random_forest, 'Random Forest')
```

```python
# 6. Gradient Boosting
gradient_boosting = GradientBoostingClassifier()
gradient_boosting.fit(X_train, y_train)
y_pred_gradient_boosting = gradient_boosting.predict(X_test)
print("Gradient Boosting")
print("Accuracy:", accuracy_score(y_test, y_pred_gradient_boosting))
print(classification_report(y_test, y_pred_gradient_boosting))
```

```python
plot_confusion_matrix(y_test, y_pred_gradient_boosting, 'Gradient Boosting')
```

```python
accuracies = {
    'SVM': accuracy_score(y_test, y_pred_svm),
    'Random Forest': accuracy_score(y_test, y_pred_random_forest),
    'Gradient Boosting': accuracy_score(y_test, y_pred_gradient_boosting)
}
```

```python
# Bar plot for accuracy
plt.figure(figsize=(10, 6))
plt.bar(accuracies.keys(), accuracies.values(), color=['blue', 'orange', 'green'])
plt.title('Model Accuracy Comparison')
plt.ylabel('Accuracy')
plt.ylim(0, 1)  # Set y-axis limits from 0 to 1
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
# Function to plot ROC curve
def plot_roc_curve(y_test, y_pred_prob, model_name):
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob, pos_label='malware')  # Assuming 'malware' is the positive class
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')
```

```python
# Get predicted probabilities for ROC curve
y_pred_prob_svm = svm.predict_proba(X_test)[:, 1]
y_pred_prob_random_forest = random_forest.predict_proba(X_test)[:, 1]
y_pred_prob_gradient_boosting = gradient_boosting.predict_proba(X_test)[:, 1]
```

```python
# Plot ROC curve
plt.figure(figsize=(10, 6))
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line
plot_roc_curve(y_test, y_pred_prob_svm, 'SVM')
plot_roc_curve(y_test, y_pred_prob_random_forest, 'Random Forest')
plot_roc_curve(y_test, y_pred_prob_gradient_boosting, 'Gradient Boosting')
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.show()
```
